```python
import pandas as pd
import numpy as np
import os
```

```python
def getPandasFromFile(path, fileName, theSkipRow):
    path = os.path.join(path , fileName)
    return pd.read_csv(path, skiprows= theSkipRow , header=None)

def SavePandasToCSV(d, path):
    d.to_csv(path)
    return "done!!"
```
