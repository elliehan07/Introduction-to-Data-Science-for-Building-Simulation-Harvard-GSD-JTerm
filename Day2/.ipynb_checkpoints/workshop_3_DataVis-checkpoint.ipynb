### 2. Data Visualization 

```python
def VisLinePlot(xData, title = 'title', YLable="YLable" , path=""):
    plt.figure(figsize=(20,10))
    yTime = pd.date_range('1/1/2016', periods=8760, freq='H')
    plt.plot(yTime,xData)
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel(YLable)
    plt.legend(loc='best')
    plt.axhline(13, color='r', label=r'True $\beta_1$')
    plt.axhline(20, color='r', label=r'True $\beta_1$')
    plt.axhspan(13, 20, facecolor='0.5', alpha=0.5,label=r'True $\beta_1$')
    if path != "":
        plt.savefig(path)
    plt.show()
    return plt
```

```python
xData = theDf["DB_temp"]
title = 'NV potential days'

thePlt = VisLinePlot(xData, title )
thePlt = VisLinePlot(xData, title, path="img/DB_temp.png")
# thePlt.
```
