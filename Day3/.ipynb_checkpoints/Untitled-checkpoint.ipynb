```python
import pandas as pd
import numpy as np
import os
import datetime
from dateutil.parser import parse
import matplotlib
import matplotlib.pyplot as plt
from __future__ import print_function
import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import MONDAY
from matplotlib.finance import quotes_historical_yahoo_ochl
from matplotlib.dates import MonthLocator, WeekdayLocator, DateFormatter
%matplotlib inline
```

```python
from os import walk
mypath = 'meter'
f = []
for (dirpath, dirnames, filenames) in walk(mypath):
    f.extend(filenames)
    break
```

```python
dfs = []
for i in f:
    fn = mypath + '/' + i
    dfs.append(pd.read_csv(fn))
```

```python
def GetMonthFromHours(data):
    lenForMonth = (len(data)/12)
    month = []
    count = 0
    theSum = 0
    for d in data:
        theSum+=d
        count +=1
        if count == lenForMonth:
            count = 0
            month.append(theSum/lenForMonth)
    return month
theData=[]
for df in dfs:
    Electricity = (df['Electricity:Facility [J](Hourly)'])
    Heating = (df['DistrictHeating:Facility [J](Hourly)'])
    d = []
    d.append(GetMonthFromHours(Electricity.tolist()))
    d.append(GetMonthFromHours(Heating.tolist()))
    theData.append(d)
```

```python
theData
```

```python
def VisLinePlotByAX(ax, xData, color = "purple", title = 'title', YLable="YLable" , path="", axLine1 = 0, axLine2 = 0 ):
    yTime = pd.date_range('1/1/2016', periods=12, freq='M')
    ax.plot(yTime,xData, alpha=0.5, color=color)
    ax.set_title(title)
    ax.set_xlabel('Year')
    ax.set_ylabel(YLable)
    ax.legend(loc='best')
    if(axLine1 != 0 and axLine2 != 0 ):
        ax.set_axhline(axLine1, color='r', label=r'True $\beta_1$')
        ax.set_axhline(axLine2, color='r', label=r'True $\beta_1$')
        ax.set_axhspan(axLine1, axLine2, facecolor='0.5', alpha=0.5,label=r'True $\beta_1$')
```

```python
d1 = theData[0][0]
d2 = theData[0][1]
# d3 = theData["Dew_Point"]
# path = "img/multiPlot"

print(len(d1))
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(1, 1, 1)
VisLinePlotByAX(ax, d1, color="red")
VisLinePlotByAX(ax, d2, color="green")
# VisLinePlotByAX(ax, d3, color="blue")
# plt.savefig(path)
plt.tight_layout()
plt.show()
```
